# Titanic Dataset Analysis
___

### Objective

- Predict if a passenger survived or not.

### Type of Problem

- This problem is classified as a supervised and offline.

**Dataset:** https://www.kaggle.com/hesh97/titanicdataset-traincsv# 


# Libraries
**_____**

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import StratifiedShuffleSplit
```

# Splitting Data
**_____**

```python
titanic_df = pd.read_csv('train.csv')

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_indx, test_indx in split.split(titanic_df, titanic_df['Sex']):
    train_set = titanic_df.loc[train_indx, :] # train_indx is an array of all the indexes that must be in the train set
    test_set = titanic_df.loc[test_indx, :] # test_indx is an array of all the indexes that must be in the test set
    
# train_set['Sex'].value_counts() / len(train_set) # male: 0.647472, female: 0.352528
# test_set['Sex'].value_counts() / len(test_set) # male: 0.648045, female: 0:351955
```

# Exploring Data
**_____**

### Data Information
- Total rows: 890
- Total columns: 12

### Data Values
- PassengerId: int64 | 0 missing
- Survived: int64 | 0 missing
- Pclass: int64 | 0 missing
- Name: string | 0 missing
- Sex: int64 | 0 missing
- Age: float64 | 176 missing
- SibSp: int64 | 0 missing
- Parch: int64 | 0 missing
- Ticket: string | 0 missing
- Fare: float64 | 0 missing
- Cabin: string | 686 missing
- Embarked: string | 1 missing

```python
fig, axs = plt.subplots(ncols=4, figsize=(16,8))

# plot all passengers
titanic_df['Sex'].value_counts().plot(kind='pie', colors=['#ffd23f', '#3a86ff'], autopct='%1.1f%%', ax=axs[0]) 

# plot survivors only
titanic_df['Survived'].groupby(titanic_df['Sex']).sum().plot(kind='pie', colors=['#3a86ff', '#ffd23f'], autopct='%1.1f%%', ax=axs[1])

# plot all passengers by class
titanic_df['Pclass'].value_counts().plot(kind='pie', colors=['#ea3546', '#662e9b', '#43bccd'], autopct='%1.1f%%', ax=axs[2])

# plot survivors by class
titanic_df['Survived'].groupby(titanic_df['Pclass']).sum().plot(kind='pie', autopct='%1.1f%%', colors=['#662e9b', '#43bccd', '#ea3546'], ax=axs[3])

# adding titles to plots
axs[0].set_title('All Passengers')
axs[1].set_title('Survivors')
axs[2].set_title('All Classes')
axs[3].set_title('Survivors by Class')

plt.rcParams['font.size'] = 10 # changing font size
plt.rcParams['font.family'] = 'serif' # changing font family

plt.show()
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(16,8))

survivors_vals = titanic_df['Survived'].value_counts()
survivors_vals.plot(kind='bar', alpha=0.8, color=['#ef476f', '#48cae4'], ax=axs[0])

axs[0].set_title('Deceased | Survived')

plt.show()
```
