## Libraries

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import StratifiedShuffleSplit, cross_val_score, cross_val_predict, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, precision_recall_curve, roc_curve, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

## Splitting Data
**_____**

```python
titanic_df = pd.read_csv('train.csv')

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_indx, test_indx in split.split(titanic_df, titanic_df['Sex']):
    train_set = titanic_df.loc[train_indx, :] # train_indx is an array of all the indexes that must be in the train set
    test_set = titanic_df.loc[test_indx, :] # test_indx is an array of all the indexes that must be in the test set
    
# train_set['Sex'].value_counts() / len(train_set) # male: 0.647472, female: 0.352528
# test_set['Sex'].value_counts() / len(test_set) # male: 0.648045, female: 0:351955
```

## Data Cleaning
**_____**


```python
median = train_set['Age'].median() # filling age missing values with median
train_set['Age'].fillna(median, inplace=True) # fill only Null rows
train_set.dropna(subset=['Embarked'], axis=0, inplace=True) # removing 2 rows that got Null Embarked values

train_set_clean = train_set.drop('Survived', axis=1) # copy of data without target values
train_set_labels = train_set['Survived'].copy()
```

*now we only need to deal with the **Cabin** column*

```python
train_set_clean.drop(['Cabin', 'Name', 'PassengerId'], axis=1, inplace=True) # removing Cabin and Names column inplace
```

*now we need to convert the **Ticket** column to ints*

```python
train_set_clean['Ticketnum'] = train_set_clean['Ticket'].apply(lambda ticket: 1 if ticket.isnumeric() else 0)
train_set_clean.drop('Ticket', axis=1, inplace=True)
```

*need to deal with **Embarked** categorical column using OneHotEncoder and ColumnTransformer*

```python
train_numeric = list(train_set_clean.drop(['Sex', 'Embarked'], axis=1)) # all values
sex_category = ['Sex']
embarked_category = ['Embarked']

num_pipeline = Pipeline([
    ('std_scaler', StandardScaler()) # feature scaling using Standardization
])

full_pipeline = ColumnTransformer([
    ('sex_transform', OrdinalEncoder(), sex_category),
    ('embarked_transform', OneHotEncoder(), embarked_category),
    ('standardization', num_pipeline, train_numeric),
])

train_set_prepared = full_pipeline.fit_transform(train_set_clean) # data prepared, numpy.ndarray
train_set_prepared[0]
```

## Training Models
**_____**

*Because this is a **classification** problem, we are going to be using binary classifier first*

### Binary Classifier 
**___**

```python
sgd_cls = SGDClassifier(random_state=42)
sgd_cls.fit(train_set_prepared, train_set_labels) # training the model

random_pass = train_set_prepared[0] 
sgd_cls.predict([random_pass]) # wrong with this passenger

cross_val_res = cross_val_score(sgd_cls, train_set_prepared, train_set_labels, cv=3, scoring='accuracy') # score 1
cross_val_pred_res = cross_val_predict(sgd_cls, train_set_prepared, train_set_labels, cv=3) # use it for confusion matrix
binary_scores = cross_val_predict(sgd_cls, train_set_prepared, train_set_labels, cv=3, method='decision_function')

res_matrix = confusion_matrix(train_set_labels, cross_val_pred_res)
precision = precision_score(train_set_labels, cross_val_pred_res)
recall = recall_score(train_set_labels, cross_val_pred_res)

f1_res = f1_score(train_set_labels, cross_val_pred_res)
fpr, tpr, thresholds = roc_curve(train_set_labels, binary_scores)
res_matrix
```

### Performance Evaluation

```python
print("\n")
print("Cross Validation 3 Folds -->")
for res in cross_val_res:
      print(f"{round((res * 100), 2)}%")
print("\n")
print("Confusion Matrix -->")
print(f"Classified Correctly: {res_matrix[0][0] + res_matrix[1][1]}")
print(f"Classified Wrongly: {res_matrix[0][1] + res_matrix[1][0]}" )
print(f"{round(((res_matrix[0][0] + res_matrix[1][1])/(res_matrix[0][0] + res_matrix[1][1] + res_matrix[0][1] + res_matrix[1][0])*100), 2)} %")
print("\n")
print(f"Precision: {round(precision, 3)*100}%")
print(f"Recall: {round(recall, 3)*100}%")
print(f"F1 Score: {round(f1_res, 3)*100}%")
```

Precision: **HIGHER** when it claims an image represents a 5 

Recall: **HIGHER** when detecting all 5's in a group of images

F1: **COMBINATED** measure of *Precision* and *Recall* 

```python
precisions, recalls, thresholds = precision_recall_curve(train_set_labels, binary_scores) # values for the plot
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds): 
    plt.plot(thresholds, precisions[:-1], 'b--', label='Precision')
    plt.plot(thresholds, recalls[:-1], 'g--', label='Recall')
    
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
plt.legend()
plt.show()
```

The *Precision/Recall Trade-off* depends on the type of project you're working on

```python
plt.plot(precisions, recalls, 'r--') # plotting precision and recall results as 'x' and 'y' values
plt.xlabel('recall')
plt.ylabel('precision')
# the closer to the top right, the better
plt.show()
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')

plot_roc_curve(fpr, tpr) # the closest to top left, the better

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate (Recall)')
plt.show()
```
